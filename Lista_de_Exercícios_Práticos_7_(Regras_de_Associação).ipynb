```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Online Retail.csv')
df.head()
```

```python
aus = (df[df['Country'] =="Australia"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

hk = basket = (df[df['Country'] =="Hong Kong"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

italia = basket = (df[df['Country'] =="Italy"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))

uk = basket = (df[df['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1

aus_sets = aus.applymap(encode_units)
hk_sets = hk.applymap(encode_units)
italia_sets = italia.applymap(encode_units)
uk_sets = uk.applymap(encode_units)
```

```python
frequent_aus = apriori(aus_sets, min_support=0.05, use_colnames=True)
frequent_hk = apriori(hk_sets, min_support=0.2, use_colnames=True)
frequent_italia = apriori(italia_sets, min_support=0.05, use_colnames=True)
frequent_uk = apriori(uk_sets, min_support=0.01, use_colnames=True)
```

```python
print("Australia:\n", frequent_aus)
print("\nHong Kong:\n", frequent_hk)
print("\nItália:\n", frequent_italia)
print("\nReino Unido:\n", frequent_uk)
```

```python
rules_aus = association_rules(frequent_aus, metric="lift", min_threshold=1)
rules_hk = association_rules(frequent_hk, metric="lift", min_threshold=1)
rules_italia = association_rules(frequent_italia, metric="lift", min_threshold=1)
rules_uk = association_rules(frequent_uk, metric="lift", min_threshold=1)
```

```python
print("Australia:\n", rules_aus.shape)
print("\nHong Kong:\n", rules_hk.shape)
print("\nItália:\n", rules_italia.shape)
print("\nReino Unido:\n", rules_uk.shape)
```

```python
actual_aus = rules_aus[ (rules_aus['lift'] >= 1) & (rules_aus['confidence'] >= 0.2) ]
actual_hk = rules_hk[ (rules_hk['lift'] >= 1) & (rules_hk['confidence'] >= 0.2) ]
actual_italia = rules_italia[ (rules_italia['lift'] >= 1) & (rules_italia['confidence'] >= 0.2) ]
actual_uk = rules_uk[ (rules_uk['lift'] >= 1) & (rules_uk['confidence'] >= 0.2) ]
```

```python
print("Australiag:\n", actual_aus.shape)
print("\nHong Kong:\n", actual_hk.shape)
print("\nItália:\n", actual_italia.shape)
print("\nReino Unido:\n", actual_uk.shape)
```
