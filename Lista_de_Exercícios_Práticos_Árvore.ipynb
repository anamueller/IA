```python
from sklearn import tree
from sklearn.datasets import load_iris
import graphviz
```

```python
#3 espécies de flor Iris;
#Iris Versicolor
#Iris Setosa
#Iris Virginica
```

Há diferença entre as árvores?
Sim, o parâmetro max_depth faz com que a 'altura' da árvore mude, quanto maior o max depth mais folhas a árvore possui. O min_samples faz com que a árvore fique mais simples a medida que o min_samples aumenta. O splitter faz com que a árvore escolha os melhores atributos e consequentemente temos uma árvore mais simples, e o random ela escolhe um atributo aleatório e parte dai, tendo uma árvore mais complexa. O parâmetro criterion define a importância de cada atributo, porém não notei diferença entre as árvores criadas com a mudança desse parâmetro.

#Árvore Inicial com Tutorial

```python
iris = load_iris()
x, y = iris.data, iris.target
arv = tree.DecisionTreeClassifier()
arv = arv.fit(x, y)
```

```python
tree.plot_tree(arv)
```

```python
dot_data = tree.export_graphviz(arv, out_file=None)
graph = graphviz.Source(dot_data) 
graph.render("iris")
```

```python
dot_data = tree.export_graphviz(arv, out_file=None, 
...                      feature_names=iris.feature_names,  
...                      class_names=iris.target_names,  
...                      filled=True, rounded=True,  
...                      special_characters=True)
```

```python
grafico = graphviz.Source(dot_data)
grafico
```
